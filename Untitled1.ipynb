```python
import os
import cv2
import numpy as np

# Initialize background subtractor
bg_subtractor = cv2.createBackgroundSubtractorMOG2()

# Function to preprocess frame
def preprocess_frame(frame):
    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(gray, (7, 7), 0)
    
    return blurred

# Function to detect hand and fingertips
def detect_hand_and_fingertips(frame):
    # Preprocess frame
    blurred = preprocess_frame(frame)

    # Apply background subtraction
    fg_mask = bg_subtractor.apply(blurred)

    # Apply thresholding
    _, thresh = cv2.threshold(fg_mask, 100, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Initialize variables for hand contour and fingertips
    hand_contour = None
    fingertips = []

    # Draw contours and find fingertips
    for contour in contours:
        # Find the largest contour (presumably the hand)
        if hand_contour is None or cv2.contourArea(contour) > cv2.contourArea(hand_contour):
            hand_contour = contour

    # Check if a hand contour was found
    if hand_contour is not None:
        # Draw hand contour
        cv2.drawContours(frame, [hand_contour], -1, (0, 255, 0), 2)

        # Simplify hand contour
        epsilon = 0.03 * cv2.arcLength(hand_contour, True)
        approx = cv2.approxPolyDP(hand_contour, epsilon, True)

        # Detect convex hull
        try:
            hull = cv2.convexHull(approx, returnPoints=False)
            if hull is not None and len(hull) > 3:
                defects = cv2.convexityDefects(approx, hull)
                if defects is not None:
                    for i in range(defects.shape[0]):
                        s, e, f, _ = defects[i, 0]
                        far = tuple(approx[f][0])
                        fingertips.append(far)
                        cv2.circle(frame, far, 5, (0, 0, 255), -1)
        except cv2.error as e:
            print(f"Error in detecting hand and fingertips: {e}")
            pass

    return frame

# Function to process video
def process_video(video_path):
    # Open video capture
    cap = cv2.VideoCapture(video_path)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("No frame read. Exiting loop.")
            break

        # Detect hand and fingertips
        processed_frame = detect_hand_and_fingertips(frame)

        # Display the frame with hand contour and fingertips
        cv2.imshow("Hand Detection", processed_frame)

        if cv2.waitKey(30) & 0xFF == ord('q'):
            break

    # Release video capture
    cap.release()
    cv2.destroyAllWindows()

# Process videos in the "Handvid" folder
video_dir = 'Handvid'
for root, dirs, files in os.walk(video_dir):
    for file in files:
        if file.endswith(('.mp4', '.avi')):
            video_path = os.path.join(root, file)
            print(f"Processing video: {video_path}")
            process_video(video_path)
```

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands model
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.5)

# Function to detect hand landmarks and draw blue dots
def detect_hand_landmarks(frame):
    # Convert BGR to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Process the frame to get hand landmarks
    results = hands.process(rgb_frame)
    
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            for landmark in hand_landmarks.landmark:
                # Extract landmark positions
                height, width, _ = frame.shape
                cx, cy = int(landmark.x * width), int(landmark.y * height)
                
                # Draw blue dots on landmarks
                cv2.circle(frame, (cx, cy), 5, (255, 0, 0), -1)
    
    return frame

# Function to calculate distance between thumb and index finger
def calculate_distance(hand_landmarks):
    thumb_tip = (hand_landmarks.landmark[4].x, hand_landmarks.landmark[4].y)
    index_tip = (hand_landmarks.landmark[8].x, hand_landmarks.landmark[8].y)
    distance = ((thumb_tip[0] - index_tip[0]) ** 2 + (thumb_tip[1] - index_tip[1]) ** 2) ** 0.5
    return distance

# Function to zoom in or out based on thumb and index finger distance
def zoom_frame(frame, distance):
    zoom_factor = 1.0 + (1.0 - distance) * 0.1  # Adjust the factor based on thumb and index finger distance
    zoomed_frame = cv2.resize(frame, None, fx=zoom_factor, fy=zoom_factor)
    return zoomed_frame

# Main function to process video feed
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Detect hand landmarks and draw blue dots
        frame_with_dots = detect_hand_landmarks(frame)
        
        # Calculate thumb and index finger distance
        # and zoom in or out based on the distance
        if hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)).multi_hand_landmarks:
            distance = calculate_distance(hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)).multi_hand_landmarks[0])
            zoomed_frame = zoom_frame(frame_with_dots, distance)
            cv2.imshow("Hand Zoom", zoomed_frame)
        else:
            cv2.imshow("Hand Zoom", frame_with_dots)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()

# Process video
video_path = "Handvid"
process_video(video_path)
```
